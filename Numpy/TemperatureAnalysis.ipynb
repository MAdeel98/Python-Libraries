```python
# Days above 35°C
```

```python
days = np.where(temp >35)[0] +1
for day in days :
    print(f" {temp[day-1]} on {day}th day")
```

```python
cool=np.where(temp<32)[0]+1
for cools in cool:
    print(f"day :{cools} and {temp[cools-1] } degre C")
```

Find all days where the temperature was between 33°C and 38°C (inclusive).

```python
temp
```

```python
days=np.where((temp>=33) & (temp<=38))[0]
for i in days:
    print(f"Day {i+1} : {temp[i]} *C ")
```

```python
condition=[
    temp<32,
    (temp>=32) &(temp<=37),
    temp>37
]
choices=["cold","normal","hot"]
weather=np.select(condition,choices,default="unknown")
print(weather)
```

```python
for i , d in enumerate(temp,start=1):
    if d <=32:
        status="cool"
    elif d <= 38 :
        status="normal"
    elif d > 38:
        status="hot"
    print(f"day {i} : {d} - {status}")
```
